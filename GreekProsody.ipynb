## Scansion métrique avec Anna Conser's `greek_scansion`

- GitHub : https://github.com/aconser/greek_scansion
- Description : *Tools for analyzing the prosody of Ancient Greek poetry and prose.*

````python
import sys
sys.path.append('./greek_scansion/')
````

````python
from greek_scansion import scan_line, syllabify
````

````python
for line in lines[:1]:
    
    # `syllabify` sert pour diviser une ligne en syllabes
    syllables = syllabify(line)
    
    # `scan_line` sert pour déterminer les longueurs métriques de chaque syllabe
    scansion = scan_line(line)
    
    print("Line: {}".format(line))
    print("Syllables: {}".format(syllables))
    print("Scansion: {}".format(scansion))
    print("*" * 50)
````

````python
# voilà une façon un peu plus elengate d'afficher la même information
for line in lines[:1]:
    
    print(line)
    syllables = syllabify(line)
    scansion = scan_line(line)
    
    for syllable, syllable_length in zip(syllables, scansion):
        print(f'{syllable}: {syllable_length}')
````

### Identifier les longueurs des syllabes

La librairie Python `hexameter` utilise les symboles suivants pour représenter les longueurs des syllabes :

```python
# excerpt from source code
# file https://github.com/epilanthanomai/hexameter/blob/master/hexameter.py

LONG = '+'
SHORT = '-'
INDETERMINATE = '?' # could be long or short
LONG_CORREPTION = 'C' # long and positionally candidate for correption
INDETERMINATE_CORREPTION = 'c' # indeterminate length but positionally
                               # candidate for correption if long
SHORT_SYNIZESIS = ',' # short and positionally candidate for synizesis

# these three are not used in input, but are used in output analysis:
SKIPPED = '.'
FOOT = '|'
```

La methode `_local_metrical_analysis` donne comment résultat une lettre (consonne/voyelle) ou grouppe de lettres, et son longueur métrique (ou "" si la lettre est une consonne)

````python
metrical_lengths = _local_metrical_analysis(lines[2])
````

````python
def classify_hexameter(scansion):
    
    # the last foot can be anything so
    # we don't consider it
    feet = scansion.split("|")[:-1]
    class_label = ""
    
    for foot in feet:
        
        # it's a spondee
        if foot == "--":
            class_label+="s"
        # it's a dactylic
        else:
            class_label+="d"
    
    return class_label
        
````

## Analyse des décasyllabes en Anglais

On va ouvrir un fichier avec un morceau de text du *Paradise Lost* par Milton.

````python
with open('data/MiltonPL1.txt', 'r') as inputfile:
    PL_text = inputfile.read()
````

````python
print(PL_text)
````

````python
syllables_by_position = {}

for n in range(1,14):
    syllables_by_position[n] = set()
````

````python
syllables_by_position
````

````python
for line in PL_text.split('\n'):

    columns = line.split('\t')
    
    if len(columns) <= 1 :
        continue
    
    for column_n, column in enumerate(columns):
        
        syllable = column.lower()
        
        syllables_by_position[column_n + 1].add(syllable)
````

````python
for position in syllables_by_position:
    print(f'{position}: {len(syllables_by_position[position])}')
````
